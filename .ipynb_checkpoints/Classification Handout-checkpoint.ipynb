```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import math
```

# Motivation: Why do we need classification methods at all? 

```python
x = np.array([0.1,0.55,0.3,0.4,0.7,0.3,0.8,0.6,0.9]).reshape((-1,1))
y = np.array([0,1,0,0,1,0,1,1,1])

model = LinearRegression()
model.fit(x, y)
df= pd.DataFrame(x,columns=['size'])
df['class'] = y
ax = sns.scatterplot(x=df['size'],y=df['class'])

# Adds straight line to represent the given hypothesis
X_plot = np.linspace(0, 1, 100)
Y_plot = model.coef_*X_plot+model.intercept_
plt.plot(X_plot, Y_plot, color='r')
plt.axvline(x=0.4949494, color='r', linestyle='dashed')
sns.despine()
```

What happens when adding extra data?

```python
x = np.append(x,4).reshape((-1,1))
y = np.append(y,1)

df= pd.DataFrame(x,columns=['size'])
df['class'] = y

model2 = LinearRegression()
model2.fit(x,y)

ax = sns.scatterplot(x=df['size'],y=df['class'])
X_plot = np.linspace(0, 4.2, 100)
Y_plot2 = model2.coef_*X_plot+model2.intercept_
plt.plot(X_plot, Y_plot, color='r')
plt.plot(X_plot, Y_plot2, color='b')
plt.axvline(x=0.381818, color='b', linestyle='dashed')
plt.axvline(x=0.4949494, color='r', linestyle='dashed')
sns.despine()
```

To attempt classification, one method is to use linear regression and map all predictions greater than 0.5 as a 1 and all less than 0.5 as a 0. However, this method doesn't work well because classification is not actually a linear function.

# Logistic Regression Model

Recall we are after outputs $0\leq h_{\theta}(x) \leq 1$

To model our logistic regression, we will take the original linear regression model <br><br>

<center>$h_{\theta}(x) = \theta^{T}x$</center>

and amend it by the addition of a new parameter $g$ to give <br><br>

<center>$h_{\theta}(x) = g(\theta^{T}x$) </center>  <br><br>

<center>where $g_{z} = \frac{1}{1+\exp^{-z}}$</center>

This is known as the **Sigmoid** or **Logistic function.** Our model (hypothesis) then becomes: <br><br>

<center> $h_{\theta}(x) = \frac{1}{1+\exp^{-{\theta}^T x}}$</center> 



```python
x = np.linspace(-10, 10, 100) 
z = 1/(1 + np.exp(-x)) 
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
  
plt.show() 
```

Note how at high and low $x$ the sigmoid satifies our output requirement of $0\leq h_{\theta}(x) \leq 1$.

We interpret the output of our model as <br><br>
<center> $h_{\theta}(x) = P(y=1|x;\theta)$</center> <br>
Or in words: the probability that y=1, given x, parameterised by $\theta$. Similarly then, for a binary classification problem:<br><br>
<center> $P(y=0|x;\theta) = 1 - P(y=1|x;\theta)$</center> 

# Decision Boundaries

Suppose we predict  $y=1$ if $h_{\theta}(x)\geq0.5$. This will be true for $\theta^{T}x \geq 0$<br> <br>
Similarly, we will predict $y=0$ if $h_{\theta}(x)\le0.5$. This will be true for $\theta^{T}x \le 0$

For an example hypothesis

 <center>$h_{\theta}(x) = g(\theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2}$) </center>  <br>where:

$$ \theta = \begin{bmatrix} -3  \\   1 \\  1 \end{bmatrix},$$

using the above inequality it is easy to show that the decision boundary would be equal to the line <br> <br>

<center>$ x_{1} + x_{2} = 3 $. </center>  <br>

**Note** the decision boundary is a property of the hypothesis including $\theta$ values, not of the data set. We will discuss how to fit $\theta$ below. 


Like in normal linear regression, we can also generate non-linear decision boundaries using logistic regresssion. To do so, we would need to use some higher order terms within our hypothesis <br><br>

<center>$h_{\theta}(x) = g(\theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2} +.  \theta_{3}x_{1}^{2} +  \theta_{4}x_{2}^{2}$) </center>  <br>

# Cost Function

In linear regression use the sum of squares cost function to penalise our model during training: <br><br>

<center>$J(\theta) = \frac{1}{2m}{\sum^{m}_{i=1}{Cost}(h_{\theta}(x^{(i)}),y^{(i)}))}$</center> <br>

where

<center>$\textrm{Cost}(h_{\theta}(x),y)= (h_{\theta}(x^{(i)})-y^{(i)})^{2},$</center> <br>


over a training set of $m$ examples: $\{(x^{(1)},y^{(1)}), (x^{(2)},y^{(2)}) ..., (x^{(m)},y^{(m)})\}$ with <br><br> <center>$x \in \begin{bmatrix} x_{0}  \\   x_{1} \\  ... \\ x_{n} \end{bmatrix}$.</center>

However, inputting our hypothesis defined above<br><br>

<center> $h_{\theta}(x) = \frac{1}{1+\exp^{-{\theta}^T x}}$</center> <br><br>

creates a non-convex $J(\theta)$. 

Or new cost function will intead be: 

<center>$\textrm{Cost}(h_{\theta}(x),y)=   \begin{cases}
    -\log(h_{\theta}(x))       & \quad \text{if } y=1\\
    -\log(1-h_\theta(x))  & \quad \text{if } y=0 
  \end{cases}$

```python
x = np.linspace(0, 1, 100) 
z = -np.log(x)
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Cost") 
plt.title('y=1')
  
plt.show() 
```

```python
x = np.linspace(0, 1, 100) 
z = -np.log(1-x)
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Cost") 
plt.title('y=0')
  
plt.show() 
```

an equivelent and more compact of writing this is: <br><br>
<center>$\textrm{Cost}(h_{\theta}(x),y)= -y\log(h_{\theta}(x))-(1-y)\log(1-h_{\theta}(x))$

giving:

<center>$J(\theta) = -\frac{1}{2m}{\sum^{m}_{i=1}}y\log(h_{\theta}(x))+(1-y)\log(1-h_{\theta}(x))$</center> <br>

and with the regularization term we have: <br>

<center>$J(\theta) = [\frac{1}{m}{\sum^{m}_{i=1}}-y^{(i)}\log(h_{\theta}(x^{(i)})) - (1-y^{(i)})\log(1-h_{\theta}(x^{(i)}))] + \frac{\lambda}{2m}\sum^{n}_{j=1}\theta_{j}^{2}  $</center> 

<br>

where m is the number of training examples, and n is the number of parameters (where $\theta_{0}$ is the parameter for the bias term).

# Partial derivatives of the cost function

Partial derivative of the regularized cost function:<br>

for j=0 (i.e., for the bias term): <br>

<center>$  \frac{\delta J(\theta)}{\delta \theta_{j}} = \frac{1}{m} {\sum^{m}_{i=1}} (h_{\theta}(x^{(i)} - y^{(i)})x_{j}^{(i)})        $<center>

<br>
    
and for j $\ge$ 1 (i.e. for the rest of the features): <br>
    
<center>$  (\frac{\delta J(\theta)}{\delta \theta_{j}} = \frac{1}{m} {\sum^{m}_{i=1}} (h_{\theta}(x^{(i)} - y^{(i)})x_{j}^{(i)})) + \frac{\lambda}{m}\theta_{j}        $<center>

# Predicting the class

I.e.: <br>
if $hyp_{\theta}(x^{(i)}) \ge 0.5, y_{pred}=1$ <br>
if $hyp_{\theta}(x^{(i)}) < 0.5, y_{pred}=0$ <br>
